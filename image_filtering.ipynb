# Filtros suaves para imagenes

En este notebook se realiza el filtrado de 4 imagenes. El proceso de filtrado es el mismo para cada gráfica. Este consiste en realizar una convolución con dos funciones, de acuerdo al tipo de filtro. Para filtros en frecuencias bajas se hace uso de una función exponencial con los suficientes datos para suavizar la gráfica. Para filtros en frecuencias altas se hace uso de la función exponencial restada a la unidad. Una vez definidas estas funciones se genera la convolución con la imagen deseada. A continuación se presenta el proceso llevado a cabo y los resultados.

```python
%pylab inline
import matplotlib.pyplot as plt
import numpy as np
from numpy import pi
from scipy.fftpack import fft, fftfreq
from scipy import fftpack
```

## Imagen 1: full_moon

Cargar la imagen y realizar la convolución con el filtro de frecuencias bajas.

```python
# read image
img = plt.imread('full_moon.jpg')
print shape(img)

# prepare an 1-D Gaussian convolution kernel
t = np.linspace(-10, 10, 20)
bump = np.exp(-0.05*t**2)
bump = bump / np.trapz(bump) # normalize the integral to 1

# make a 2-D kernel out of it
kernel = bump[:, np.newaxis] * bump[np.newaxis, :]
print img.shape[:2]

# padded fourier transform, with the same shape as the image
kernel_ft = fftpack.fft2(kernel, shape=img.shape[:2], axes=(0, 1))

# convolve
img_ft = fftpack.fft2(img, axes=(0, 1))
img2_ft = kernel_ft[:, :, np.newaxis] * img_ft
img2 = fftpack.ifft2(img2_ft, axes=(0, 1)).real
```

```python
fig = plt.figure(1, figsize=(9.5,9))
plt.imshow(img)
```

Filtro suave para frecuencias bajas:

```python
fig = plt.figure(1, figsize=(9.5,9))
imshow(img2, cmap = cm.Greys_r)
plt.title('Filtro suave para frecuencias bajas', fontsize=12)
```

Realizar la convolución para filtros de alta frecuencia:

```python
# prepare an 1-D Gaussian convolution kernel
t = np.linspace(-10, 10, 20)
bump2 = 1 - np.exp(-0.05*t**2)
bump2 = bump2 / np.trapz(bump2) # normalize the integral to 1

# make a 2-D kernel out of it
kernel2 = bump2[:, np.newaxis] * bump2[np.newaxis, :]
print img.shape[:2]

# padded fourier transform, with the same shape as the image
kernel2_ft = fftpack.fft2(kernel2, shape=img.shape[:2], axes=(0, 1))

# convolve
img_ft = fftpack.fft2(img, axes=(0, 1))
img3_ft = kernel2_ft[:, :, np.newaxis] * img_ft
img3 = fftpack.ifft2(img3_ft, axes=(0, 1)).real
```

```python
fig = plt.figure(1, figsize=(9.5,9))
imshow(img3, cmap = cm.Greys_r)
plt.title('Filtro suave para frecuencias altas', fontsize=12)
```

Para ambos filtros se nota en la región dentro de la luna mucho ruido de todo el rango de frecuencias. Pero en los bordes de la luna se pueden notar diferencias. El filtro de bajas frecuencias reduce el ruido en los bordes, haciendo que esté se vea más suavizado al aplicar este filtro. En contraste para altas frecuencias, el ruido en los bordes es mayor, por lo cual se nota un borde difuminado.
