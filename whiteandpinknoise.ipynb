# White noise Vs. Pink noise

En este notebook se realiza la transformada de fourier de una señal grabada por la herramienta SOX. A partir de la transformada se convierte la señal en white noise y en pink noise. Además se crean los archivos de voz generados al convertir la señal en los dos distintos ruidos. Por último se realiza una gráfica superpuesta de la señal con ambos ruidos.

```python
%pylab inline
import matplotlib.pyplot as plt
import numpy as np
from numpy import pi
```

```python
from scipy.fftpack import fft, fftfreq
from scipy import fftpack
```

```python
import scikits.audiolab as audio
from scipy.fftpack import fft, fftfreq
from __future__ import division
```

Para comenzar debemos leer la señal de sonido "voice.wav" grabada. A partir de estas señales se debe crear dos arreglos, uno de la señal y uno de tiempo.

```python
input_signal, sampling_rate, enc = audio.wavread("voice.wav")
print (input_signal[0:10]), sampling_rate, enc
```

```python
time_array = arange(0, len(input_signal)/float(sampling_rate), 1/float(sampling_rate))
print len(time_array), len(input_signal)
```

A continuación se presenta la gráfica de la señal en función del tiempo.

```python
n_points  = len(input_signal)
plot(time_array, input_signal)
xlabel("time(s)", fontsize=20)
ylabel("Amplitude", fontsize=20)
```

## FFT y espectro de potencias de la grabación

Procedemos a realizar la transformada de fourier de la señal, teniendo en cuenta la normalización debida al tamaño del arreglo. Una vez realizada, la transformada de fourier, calculamos el espectro de potencias de la señal. Por último, se vuelve a la señal realizando la transformada inversa de Fourier.

```python
fft_input_signal = fft(input_signal) / n_points 
freq = fftfreq(n_points, 1) 
plt.plot(freq,np.abs(fft_input_signal))
plt.xlim(-0.05,0.05)
xlabel("frequency(Hz)", fontsize=20)
ylabel("Amplitude of FFT", fontsize=20)
```

Transformada de Fourier de la señal.

```python
half_n = np.ceil(n_points/2.0)
fft_input_signal_half = (2.0) * fft_input_signal[:half_n]
freq_half = freq[:half_n]
```

```python
scatter(freq_half, log10(np.absolute(fft_input_signal_half)**2))
xlabel("frequency(Hz)", fontsize=20)
ylabel("Power Amplitude", fontsize=20)
```

Espectro de potencias de la señal.

```python
clean_f = ifft(fft_input_signal*n_points) 
plot(clean_f)
xlabel("time", fontsize=20)
ylabel("Amplitude", fontsize=20)
```

Transformada inversa de Fourier de la señal.

### White Noise

En el espectro de potencias, el ruido blanco se caracteriza por ser una constante. Esto se consigue, al normalizar la señal. Al realizar la normalización se obtiene la conversión a ruido blanco de la señal. Además se realiza la gráfica del espectro de potencias, para confirmar que sea ruido blanco.

```python
fft_input_signal_white_noise = fft_input_signal / np.absolute(fft_input_signal)
clean_white_noise = ifft(fft_input_signal_white_noise) 
plot(clean_white_noise)
xlabel("time", fontsize=20)
ylabel("Amplitude", fontsize=20)
output_file = audio.Sndfile("whitenoise.wav", 'w', audio.Format('wav'), 1, sampling_rate)
output_file.write_frames(clean_white_noise)
output_file.close()
```

Señal convertida a ruido blanco.

```python
half_n = np.ceil(n_points/2.0)
fft_input_signal_half_white_noise = (2.0) * fft_input_signal_white_noise[:half_n]
freq_half_norm = freq[:half_n]
scatter(freq_half_norm, log10(np.absolute(fft_input_signal_half_white_noise)**2)) 
xlabel("frequency (Hz)", fontsize=20)
ylabel("Power Amplitude", fontsize=20)
```

Espectro de potencias del ruido blanco.

### Pink noise

En el espectro de potencias, el ruido rosa se caracteriza por ser inversamente proporcional a una potencia n de la frecuencia. Esto se consigue, al dividir la señal convertida a ruido blanco por la frecuencia. Al realizar la división se obtiene la conversión a ruido rosa de la señal. Además se realiza la gráfica del espectro de potencias, para confirmar que sea ruido rosa.

```python
fft_input_signal_pink_noise = fft_input_signal_white_noise[1:] / freq[1:]
clean_pink_noise = ifft(fft_input_signal_pink_noise) 
plot(clean_pink_noise)
xlabel("time", fontsize=20)
ylabel("Amplitude", fontsize=20)
output_file = audio.Sndfile("pinknoise.wav", 'w', audio.Format('wav'), 1, sampling_rate)
output_file.write_frames(clean_pink_noise)
output_file.close()
```

Señal convertida en ruido rosa.

```python
half_n = np.ceil(n_points/2.0)
fft_input_signal_half_pink_noise = (2.0) * fft_input_signal_pink_noise[:half_n]
freq_half_norm = freq[:half_n]
scatter(freq_half_norm, log10(np.absolute(fft_input_signal_half_pink_noise)**2)) 
xlabel("time", fontsize=20)
ylabel("Amplitude", fontsize=20)
```

Espectro de potencias del ruido rosa.

### Comparación entre white noise y pink noise

Por último se superponen las señales convertidas para poder compararlas.

```python
plot(clean_white_noise)
plot(clean_pink_noise)
xlabel("time", fontsize=20)
ylabel("Amplitude", fontsize=20)
```

En el espectro de potencias de la señal de ruido rosa, es de notar que la amplitud de la potencia durante todo el espectro de frecuencias siempre es mayor a la amplitud de la potencia del ruido blanco. Esto se evidencia en esta última gráfica donde la amplitud del ruido rosa siempre es mayor a la amplitud del ruido blanco.
